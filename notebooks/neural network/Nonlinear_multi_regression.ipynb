```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create nonlinear feature maps for 6 input variables
n   = 500

# The ground truth intercept and weights
w_true = [0.5, 1.5, -1.0, 0.7, 1.3, 2.0, 1.0]   
sigma = 0.5

# Create Data x
x1 = np.random.uniform(-1, 1, n)
x2 = np.random.uniform(-1, 1, n)
x3 = np.random.uniform(-1, 1, n)
x4 = x1 * x2                                      
x5 = x2 ** 2                                      
x6 = np.cos(x1)                                       

# Create Matrix
X = [x1, x2, x3, x4, x5, x6]
# print(X) 
print (X[: 1])
Y = w_true[0]*x1 + w_true[1]*x2 + w_true[2]*x3 + w_true[3]*x4 + w_true[4]*x5 + w_true[5]*x6 + w_true[6] + sigma
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nonlinear regression')
plt.scatter(X[: 1], Y, marker='o')
plt.show()
```
